```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Tmall_brand.xlsx')
data
```

```python
brands = list(set(data['main_brand_name']))
```

```python
result =pd.DataFrame(columns=('brand','gmv','sales_qtty','atv'))
for brand in brands:
    tmp_data = data[data['main_brand_name'] == brand]
    sum_1 = tmp_data.sum()
    gmv = sum_1['gmv']
    sales = sum_1['sale_qtty']
    atv = gmv / sales
    result=result.append(pd.DataFrame({'brand':[brand],'gmv':[gmv],'sales_qtty':[sales],'atv':[atv]}),ignore_index=True)
```
