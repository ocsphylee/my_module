```python
f1 = pd.read_excel('Econ_data.xls')
f1
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False

fig, axes = plt.subplots(figsize=(16,9))
x = range(len(f1['年份']))
x_tic = [i*3 for i in x]
bar_1 = axes.bar(x_tic,f1['1季度'],width = 0.5, alpha = 0.8,label = '一季度')
bar_2 = axes.bar([i + 0.5 for i in x_tic],f1['2季度'],width = 0.4, alpha =0.8,label = '二季度')
bar_3 = axes.bar([i + 1 for i in x_tic],f1['3季度'],width = 0.4, alpha = 0.8,label = '三季度')
bar_4 = axes.bar([i + 1.5 for i in x_tic],f1['4季度'],width = 0.4, alpha = 0.8,label = '四季度')
plt.ylim(0,12)
plt.ylabel('GDP同比增长速度',fontsize = 15)
plt.xlabel('年份',fontsize = 15)

plt.xticks([i + 0.75 for i in x_tic],f1['年份'])

for rect in bar_1:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height+0.1, str(height), ha="center", va="bottom")
for rect in bar_2:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height+0.1, str(height), ha="center", va="bottom")    
for rect in bar_3:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height+0.1, str(height), ha="center", va="bottom")
for rect in bar_4[0:5]:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height+0.1, str(height), ha="center", va="bottom")

plt.text(-1,-1,'数据来源：国家统计局',fontsize=12)
growth = [7.8,7.3,6.9,6.7,6.9]

for a,rect in zip(growth,bar_2):
    plt.text(rect.get_x(),9,'---------')
    plt.text(rect.get_x()+0.3,9.3,a)
    plt.text(rect.get_x()+0.1,8.7,'全年GDP')
plt.text(15,9,'---------')
plt.text(15.3,9.3,6.7)
plt.text(15,8.7,'1-3季度GDP')

plt.title('2013-2018年各季度国内生产总值（GDP）情况',fontsize=20)
axes.legend(loc = 1,fontsize = 15,ncol = 4)
```

```python
f2 = pd.read_excel('growth.xls')
f2
```

```python
fig2,axes2 = plt.subplots(2,1,figsize = (16,9))
x = range(len(f2['指标']))
plt.rcParams['savefig.dpi'] = 500 #图片像素
plt.rcParams['figure.dpi'] = 500#分辨率

#图1
axes2[0].set_title('图2 规模以上工业企业主营业务收入收入累计增长（同比）',fontsize = 20)

axes2[0].plot(f2['指标'],f2['国主增'],'o-',label = '国有控股企业主营业务收入累计增长')
for a, b in zip(f2['指标'],f2['国主增']):
    axes2[0].text(a,b+0.5,b,ha = 'center',fontsize = 15)

axes2[0].plot(f2['指标'],f2['私主增'],'s-',label = '私营企业主营业务收入累计增长')
for a, b in zip(f2['指标'],f2['私主增']):
    axes2[0].text(a,b-1,b,ha = 'center',fontsize = 15)

axes2[0].plot(f2['指标'],f2['国有主增'],'^-',label = '国有企业主营业务收入累计增长')
for a, b in zip(f2['指标'],f2['国有主增']):
    axes2[0].text(a,b+0.5,b,ha = 'center',fontsize = 15)

axes2[0].grid(axis="y")
axes2[0].set_ylim(8,20)
axes2[0].legend(ncol = 3,fontsize = 15)
axes2[0].set_ylabel('累计增长率',fontsize = 15)
axes2[0].set_xlabel('月份',fontsize = 15)
axes2[0].text(-0.35,5,'数据来源：国家统计局',fontsize = 12)

#图2
axes2[1].plot(f2['指标'],f2['国利增'],'o-',label = '国有控股企业利润总额累计增长')
for a, b in zip(f2['指标'],f2['国利增']):
    axes2[1].text(a,b+3,b,ha = 'center',fontsize = 15)

axes2[1].plot(f2['指标'],f2['私利增'],'s-',label = '私营企业利润总额累计增长')
for a, b in zip(f2['指标'],f2['私利增'][0:-2]):
    axes2[1].text(a,b-4,b,ha = 'center',fontsize = 15)    
axes2[1].text(6,12,10,ha = 'center',fontsize = 15)
axes2[1].text(7,11.3,9.3,ha = 'center',fontsize = 15)

axes2[1].plot(f2['指标'],f2['国有利增'],'^-',label = '国有企业利润总额累计增长')
for a, b in zip(f2['指标'],f2['国有利增']):
    axes2[1].text(a,b,b,ha = 'left',fontsize = 15)

axes2[1].grid(axis="y")
axes2[1].set_ylim(0,45)
axes2[1].legend(ncol = 3,fontsize = 15)
axes2[1].set_ylabel('累计增长率',fontsize = 15)
axes2[1].set_xlabel('月份',fontsize = 15)
axes2[1].text(-0.35,-10,'数据来源：国家统计局',fontsize = 12)
axes2[1].set_title('图3 规模以上工业企业利润总额收入累计增长（同比）',fontsize = 20)


#fig.tight_layout()#调整整体空白
plt.subplots_adjust(wspace =0, hspace =0.5)#调整子图间距
```
