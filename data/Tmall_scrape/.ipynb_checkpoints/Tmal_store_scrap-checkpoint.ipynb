```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
import pandas as pd
from datetime import datetime
```

```python
def get_cat_dic(url):
    driver.get(url)
    time.sleep(5)
    a = BS(driver.page_source).find_all('a',{ "class": "cat-name fst-cat-name"})
    url_dic = {}
    for item in a[3:]:
        text = item.text.strip('【】')
        url = 'https:' + item['href']
        url_dic[text] = url
    return url_dic
```

```python
def get_detail(item):
    text = item.find_element_by_class_name('detail').text.split()
    name = text[0]
    price = text[1].strip('¥')
    sales = text[2].strip('月销量：')
    return name, float(price),int(sales)
```

```python
def get_item(itemlines):
    for lines in itemlines[:-2]:
        for item in lines:
            name,price,sales = get_detail(item)
    return name, float(price),float(sales)
```

```python
def get_data(cid,url,data,i=0):
    driver.get(url)
    time.sleep(5)
    itemlines = driver.find_elements_by_class_name('item4line1')
    try:
        tmp = itemlines[0]
    except:
        i +=1
        print("timeout,retring {} time".format(i))
        time.sleep(5)
        save_data(cid,url,data,i=i)
    else:
        print('succeeded in read data')
        for lines in itemlines[:-2]:
            items = lines.find_elements_by_class_name('item')
            for item in items:
                name,price,sales = get_detail(item)
                gmv = price * sales
                data.loc[len(data)]=[cid,name,price,sales,gmv]
        time.sleep(10)
```

```python
def save_data(data,path,date):
    dt = datetime.strptime(date,"%Y-%m-%d")
    data['dt'] = dt
    writer = pd.ExcelWriter(path,engine='openpyxl')
    data.to_excel(writer,sheet_name = '原始数据')
    group_data = data.groupby('cid').sum()
    group_data.loc['总计'] = group_data.sum()
    group_data['dt'] = dt
    group_data.to_excel(writer,sheet_name = '分类数据')
    writer.save()
    writer.close()
```

```python
options = webdriver.ChromeOptions()
options.add_argument('user-agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36"')
options.add_experimental_option('excludeSwitches', ['enable-automation']) 
driver = webdriver.Chrome(chrome_options=options)
```

```python
data = pd.DataFrame(columns=['cid','name','price','sales','gmv'])
for key,value in url_dic.items():
    get_data(key,value,data)
```
